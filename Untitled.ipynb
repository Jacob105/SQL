```python
def new_engine(dialect, server, database, user=None, password=None, integrated_security=True):
    if integrated_security:
        # For Windows authentication
        eng = f"{dialect}://{server}/{database}?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server"
    else:
        # For SQL Server authentication
        eng = f"{dialect}://{user}:{password}@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server"
    print(eng)
    return create_engine(eng)
```

```python
#Showing all Sales Persons, their gender,name, age and salesYTD
SalesPerson_sql ="""
SELECT A.BusinessEntityID
	,A.JobTitle
	,A.BirthDate
	,A.Gender
	,B.FirstName
	,B.MiddleName
	,B.LastName
	,C.SalesYTD
FROM HumanResources.Employee AS A
INNER JOIN Person.Person AS B
ON A.BusinessEntityID = B.BusinessEntityID
INNER JOIN Sales.SalesPerson AS C
ON A.BusinessEntityID = C.BusinessEntityID
WHERE JobTitle = 'Sales Representative'
ORDER BY SalesYTD DESC;
"""
pd.read_sql(sql=SalesPerson_sql, con=connection)
```

```python
GenderCount_sql ="""
SELECT
  JobTitle,
  COUNT(CASE WHEN Gender = 'M' THEN 1 END) AS MaleCount,
  COUNT(CASE WHEN Gender = 'F' THEN 1 END) AS FemaleCount
FROM
  HumanResources.Employee
WHERE
  JobTitle = 'Sales Representative'
GROUP BY
  JobTitle;
"""
pd.read_sql(sql=GenderCount_sql, con=connection)
```

```python
#Calculating the mean of the sales for male and femamle
SalesMean_sql ="""
SELECT Gender,
    SUM(SalesYTD) / CASE WHEN Gender = 'M' THEN 8
                         WHEN Gender = 'F' THEN 6
           ELSE 1 END AS AverageSalesYTD
FROM HumanResources.Employee AS A
INNER JOIN Sales.SalesPerson AS C 
ON A.BusinessEntityID = C.BusinessEntityID
WHERE JobTitle = 'Sales Representative'
GROUP BY Gender;
"""
pd.read_sql(sql=SalesMean_sql, con=connection)
```

```python
#Looking at the most sold items
MostSoldItems_sql ="""
SELECT TOP 10
  A.ProductID,
  A.Name,
  SUM(B.OrderQty) AS TotalSold
FROM 
  Production.Product AS A
  JOIN Sales.SalesOrderDetail AS B
  ON A.ProductID = B.ProductID 
  JOIN Sales.SalesOrderHeader AS C 
  ON B.SalesOrderID = C.SalesOrderID
GROUP BY
  A.ProductID, A.Name
ORDER BY
  TotalSold DESC;
"""
pd.read_sql(sql=MostSoldItems_sql, con=connection)
```

```python
#Sales by region
SalesRegion_sql ="""
SELECT [Group]
      ,SUM(SalesYTD) AS TotalSalesThisYear
      ,SUM(SalesLastYear) AS TotalSalesLastYear
FROM Sales.SalesTerritory
GROUP BY [Group]
ORDER BY TotalSalesThisYear DESC;
 """
pd.read_sql(sql=SalesRegion_sql, con=connection)
```

```python
#Sales Growth by region
SalesGrowth_sql ="""
SELECT [Group]
    ,SUM(SalesYTD) AS TotalSalesThisYear
    ,SUM(SalesLastYear) AS TotalSalesLastYear
    ,((SUM(SalesYTD) - SUM(SalesLastYear)) / NULLIF(SUM(SalesLastYear), 0)) * 100 AS GrowthPercentage
FROM Sales.SalesTerritory
GROUP BY [Group]
ORDER BY TotalSalesThisYear DESC;
 """
pd.read_sql(sql=SalesGrowth_sql, con=connection)
```

```python
# Sales data for this year and last year
regions = ['Europe', 'Pacific', 'North America']
sales_this_year = [13590506.0212, 5977814.9154, 33182889.0168]
sales_last_year = [5340312.9485, 2278548.9776, 25096493.5092]

# Function to format y-axis ticks as millions
def millions_formatter(x, pos):
    return f'${int(x/1e6)}M'

# Calculate growth percentage
growth_percentage = [((this_year - last_year) / last_year) * 100 if last_year != 0 else 0
                     for this_year, last_year in zip(sales_this_year, sales_last_year)]

# Plotting
bar_width = 0.35
index = range(len(regions))

fig, ax = plt.subplots(figsize=(10, 10))  # Adjust the figure size here
bar1 = ax.bar(index, sales_this_year, bar_width, label='This Year')
bar2 = ax.bar([i + bar_width for i in index], sales_last_year, bar_width, label='Last Year')

# Adding labels, title, and legend
ax.set_xlabel('Regions')
ax.set_ylabel('Total Sales (Millions USD)')
ax.set_title('Total Sales This Year and Last Year by Region')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(regions)
ax.legend()

# Adding annotations above each bar with improved positioning and formatted as millions
for i, (v_this_year, v_last_year, growth) in enumerate(zip(sales_this_year, sales_last_year, growth_percentage)):
    ax.text(i, v_this_year + 0.005 * max(sales_this_year), f'{int(v_this_year/1e6)}M\n(+{growth:.1f}%)',
            ha='center', va='bottom', color='black', fontsize=9)

for i, v in enumerate(sales_last_year):
    ax.text(i + bar_width, v + 0.02 * max(sales_last_year), f'{int(v/1e6)}M', ha='center', va='bottom', color='black', fontsize=9)

# Format y-axis ticks as millions
ax.yaxis.set_major_formatter(FuncFormatter(millions_formatter))

# Display the plot
plt.show()
```

```python
# Antal tillverkade föremål
totalt_tillverkade = 22009

# Antal trasiga föremål
antal_hela = 21316

# Andelen trasiga föremål
andelen_hela = antal_hela / totalt_tillverkade

# Skapa ett binominalt fördelat objekt
binomial_dist = sm.stats.proportion_confint(antal_hela, totalt_tillverkade, alpha=0.05, method='normal')

# Hämta konfidensintervallet
konfidensintervall = (binomial_dist[0], binomial_dist[1])

print(f"Andelen hela föremål: {andelen_hela:.4f}")
print(f"95% Konfidensintervall för andelen hela föremål: ({konfidensintervall[0]:.4f}, {konfidensintervall[1]:.4f})")
```

```python
#Show what products that is most odered
Orders_sql ="""
SELECT TOP 10 SUM(OrderQty) AS TotalOrders
	,SUM(A.ProductID) AS ItemID
	,B.Name
FROM Purchasing.PurchaseOrderDetail AS A
INNER JOIN Production.Product AS B
ON A.ProductID = B.ProductID
GROUP BY B.Name
ORDER BY TotalOrders DESC;
"""
pd.read_sql(sql=Orders_sql, con=connection)
```

```python
#Showing total scrapped items, the percentage and the reason why the item is scrapped
ScrappedItems_sql ="""
SELECT
    SUM(A.OrderQty) AS TotalOrders,
    SUM(A.ScrappedQty) AS ScrappedItems,
    ROUND(CAST(CAST(SUM(A.ScrappedQty) AS DECIMAL) 
  / NULLIF(CAST(SUM(A.OrderQty) AS DECIMAL), 0) * 100 AS DECIMAL(4, 1)), 1) AS ScrapPercent,
    B.Name
FROM Production.WorkOrder AS A
INNER JOIN Production.ScrapReason AS B 
ON A.ScrapReasonID = B.ScrapReasonID
GROUP BY B.Name
ORDER BY ScrapPercent DESC;
"""
pd.read_sql(sql=ScrappedItems_sql, con=connection)
```

```python
#Total cost of the scrapped Items
TotalScrapCost_sql ="""
SELECT SUM(TotalCost) AS TotalScrapCost
FROM (
    SELECT SUM(B.StandardCost) AS TotalCost
    FROM Production.WorkOrder AS A
    INNER JOIN Production.ProductCostHistory AS B 
	ON A.ProductID = B.ProductID
    WHERE ScrappedQty > 0
    GROUP BY A.ProductID
) AS Subquery;
"""
pd.read_sql(sql=TotalScrapCost_sql, con=connection)
```

```python
# Given values
lower_bound = 0.9662
upper_bound = 0.9708

# Calculate the mean and standard deviation from the confidence interval bounds
mean = (lower_bound + upper_bound) / 2
std_dev = (upper_bound - lower_bound) / (2 * norm.ppf(0.975))  # 0.975 corresponds to a 95% confidence interval

# Generate data points for the normal distribution within the confidence interval
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 1000)
y = norm.pdf(x, mean, std_dev)

# Plot the normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, y, where=[(lower_bound <= xi <= upper_bound) for xi in x], alpha=0.3, label='95% CI')



# Mark the beginning and end of the interval with exact numbers
plt.axvline(x=lower_bound, color='green', linestyle='--', label=f'Lower Bound ({lower_bound:.4f})')
plt.axvline(x=upper_bound, color='blue', linestyle='--', label=f'Upper Bound ({upper_bound:.4f})')

# Add labels and title
plt.title('95% Confidence Interval for Proportion')
plt.xlabel('Proportion')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
